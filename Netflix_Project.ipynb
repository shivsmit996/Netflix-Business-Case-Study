```python
def rating_barplot(data, title, height, h_lim=None):
    fig, ax = plt.subplots(1,1, figsize=(15, 7))
    if h_lim :
        ax.set_ylim(0, h_lim)
    ax.bar(data.index, data,  color="#d0d0d0", width=0.6, edgecolor='black')

    color =  ['green',  'blue',  'orange',  'red']
    span_range = [[0, 2], [3,  6], [7, 8], [9, 11]]

    for idx, sub_title in enumerate(['Little Kids', 'Older Kids', 'Teens', 'Mature']):
        ax.annotate(sub_title,
                    xy=(sum(span_range[idx])/2 ,height),
                    xytext=(0,0), textcoords='offset points',
                    va="center", ha="center",
                    color="w", fontsize=16, fontweight='bold',
                    bbox=dict(boxstyle='round4', pad=0.4, color=color[idx], alpha=0.6))
        ax.axvspan(span_range[idx][0]-0.4,span_range[idx][1]+0.4,  color=color[idx], alpha=0.1)

    ax.set_title(f'Distribution of {title} Rating', fontsize=20, fontweight='bold', position=(0.5, 1.0+0.03))
    plt.show()
```

   # Business Insights

Here are some business insights that can be drawn from the range of attributes :          

1. Type of show: Netflix's content library is dominated by movies, which account for about 70% of the titles. This is likely because movies are typically shorter and easier to produce than TV shows, and they can be more easily marketed to a global audience.

2. Release year: The release year for shows is concentrated in the range 2000-2021. This suggests that Netflix is constantly adding new content to its library, and that it is also investing in older content that may have been overlooked by other streaming services.

3. Year of adding shows: The year of adding shows is concentrated in the range 2015-2021. This suggests that Netflix has been expanding its content library rapidly in recent years, and that it is likely to continue to do so in the future.

4. Movie duration: The movie duration range of 50-150 minutes is consistent with the typical length of a feature film. This suggests that Netflix is not trying to compete with traditional TV networks by offering shorter, more episodic content

5. TV show duration: The TV show duration range of 1-3 seasons is also consistent with the typical length of a TV series. This suggests that Netflix is not trying to compete with traditional cable networks by offering longer, serialized content.

6. Countries: The fact that only 23 countries have more than 50 movie titles and only 11 countries have more than 50 TV shows suggests that Netflix is still heavily focused on the US market. However, the company is clearly expanding its international presence, and it is likely to continue to do so in the future.

7. Ratings: The 12 different kinds of ratings based on the relevance of the content for different age-groups suggest that Netflix is taking a proactive approach to ensuring that its content is accessible to a wide range of viewers.

8. Actors and directors: The number of actors and directors in the dataset suggests that Netflix is working with a large and diverse pool of creative talent. This is likely to be a key factor in the company's ability to continue to produce high-quality content.
